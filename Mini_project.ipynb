```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Matplotlib/Projects/WHR2023.csv")
df.index = range(1, len(df)+1)
df
```

## Bargraph of topmost 10 happiest countries

```python
top_ladder_score = df["Ladder score"].iloc[:10]
top_countries_name = df["Country name"].iloc[:10]

plt.figure(figsize = (13,8))
plt.title("Bar graph indicating top 10 most happiest countries by 2023")
plt.bar(
    top_countries_name,
    top_ladder_score,
    color="skyblue",
)

plt.legend(["Ladder Score"])

plt.show()
```

## World's most least happiest countries in comparision to the most happiest ones

```python
least_ladder_score = df["Ladder score"].iloc[-10:]
least_countries_name = df["Country name"].iloc[-10:]

plt.figure(figsize=(15, 8))
plt.title("Bar graph indicating top 10 least happiest countries by 2023")
plt.bar(
    least_countries_name,
    least_ladder_score
)
plt.show()

```

## Visualization of relationship between GDP and ladder score of top 10 happiest countries in the world

```python
ladder_score = df["Ladder score"].iloc[:10]
lodged_GDP = df["Logged GDP per capita"].iloc[:10]

plt.title("Relationship between top 10 country's ladder score and GDP:\n", fontsize=12, fontweight="bold")
plt.plot(
    ladder_score,
    lodged_GDP,
    marker=" ",
    linestyle="-",
    color="orange",
    linewidth=2
)
plt.xlabel("\n\nGross Domestic Product (GDP) --->\n", fontsize=15)
plt.ylabel("\nLabel Score --->\n\n", fontsize=15)
plt.xticks(fontsize=8, rotation=15)
plt.yticks(fontsize=8, rotation=-15)
plt.legend(["Ladder Score"])
plt.grid(True, linestyle=":", linewidth=0.4)
plt.show()
```

## Scatter graph

```python
ladder_score = df["Ladder score"].iloc[:10]
lodged_GDP = df["Logged GDP per capita"].iloc[:10]

plt.scatter(
    lodged_GDP,
    ladder_score,
    color="green",
    alpha = 0.6
)
plt.title("Scatter visualization in between lodged GDP and ladder score \n")
plt.ylabel("Ladder Score \n")
plt.xlabel("\n Lodged GDP")
plt.xticks(rotation=5)
plt.yticks(rotation=-5)
plt.grid(True, linestyle="-", linewidth=0.3)
plt.show()
```
